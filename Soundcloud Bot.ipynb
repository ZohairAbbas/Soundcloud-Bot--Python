```python
import time
from selenium import webdriver
```

```python
header = '''
  ___
/ ___|  ___  _   _ _ __   __| | ___| | ___  _   _  __| |
\___ \ / _ \| | | | '_ \ / _` |/ __| |/ _ \| | | |/ _` |
 ___) | (_) | |_| | | | | (_| | (__| | (_) | |_| | (_| |
|____/ \___/ \__,_|_| |_|\__,_|\___|_|\___/ \__,_|\__,_|
 ____  _
|  _ \| | __ _ _   _  ___ _ __
| |_) | |/ _` | | | |/ _ \ '__|
|  __/| | (_| | |_| |  __/ |
|_|   |_|\__,_|\__, |\___|_|
               |___/

Made by Zohair Abbas
'''
  
def print_header():
    print(header)
    
def initiate_browser():
    driver = webdriver.Chrome() 
    return driver

def search(song_name):
    search_bar = driver.find_element_by_xpath("//*[@id='content']/div/div/div[2]/div/div[1]/span/span/form/input")
    search_bar.send_keys(song_name+"\n")
    
def play_and_pause():
    driver.find_element_by_css_selector("a.snippetUXPlayButton").click()
    
def search_and_play():
    time.sleep(1)
    driver.find_element_by_css_selector("button#onetrust-accept-btn-handler").click()
    song_name = input("Enter your favourite song: ")
    search(song_name)
    time.sleep(3)
    play_and_pause()
    flag = 0
    while flag!=2:
        print(f"\n\n{'Playing' if flag == 0 else 'Paused'} {song_name.title()}, press P to {'pause' if flag == 0 else 'play'} it")
        print('or enter X to exit the program')
        
        response = input("Response: ")
        if response.lower() == 'p':
            play_and_pause()
            print_header()
            if flag == 0:
                flag = 1
            else:
                flag = 0
        elif response.lower() == 'x':
            flag = 2
        else:
            print_header()
    print("Exiting the program")
    
  
def main():
    print_header()
    global driver
    driver = initiate_browser()
    driver.get("https://www.soundcloud.com")
    search_and_play()
    driver.quit()
    print("\n\nProgram Terminated")
    
if __name__ == "__main__":
    main()
```
